# モンテカルロ法による円周率の導出

---
## 手法の解説

- 一辺の長さを r とする正方形の面積を S とし、  
その正方形の左下頂点を中心とする 半径rの４分円の面積を Q とすると  
Q / S = ( π* r * r / 4 ) / ( r * r ) = π/4  
すなわち、π = Q / S *4 という関係式が得られます。
- 正方形の「枠内」に一様に「雨」を降らせると  
正方形全体に降る雨粒の数：４分円の中に降る雨粒の数 ≒ S：Q となり  
雨粒の数が多くなれば、S：Q の比率に一致します。
-これを利用すると  
π = ４分円の中に降る雨粒の数 / 正方形全体に降る雨粒の数 * 4  
というかたちで、π の近似値を求めることができます。

---

```python
# ライブラリの読み込み
import math
import matplotlib.pyplot as plt

# 初期化
n = 1000
x = [0]*n
y = [0]*n
cnt_in = 0
cnt_out = 0

# n個のランダムな「雨粒」
for i in range(n):

  x[i] = random.random()
  y[i] = random.random()
  r = math.sqrt(x[i]**2 + y[i]**2)

  # 4分円の中に落ちた数をカウント
  if r <= 1.0:
    cnt_in += 1
  else:
    cnt_out += 1

# 円周率の近似値
Pi = cnt_in / n * 4 

# 結果の表示
print("Num：", n)
print("IN：", cnt_in)
print("OUT：", cnt_out)
print("Pi：", Pi)

# グラフの表示
plt.figure(figsize=(8, 8), dpi=80)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.scatter(x, y, marker=".", color = "blue", label = "POINT")

cir = plt.Circle((0, 0), radius=1, linewidth=2, ec="red", fc="none")
ax = plt.gca()
ax.add_patch(cir)

plt.show()
```
