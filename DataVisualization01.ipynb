# グラフ描画サンプル 

## ライブラリの読み込み

```python
# 数値計算、ビジュアライゼーションライブラリの読み込み
import numpy as np
import matplotlib.pyplot as plt
```

## 乱数を用いたグラフ

```python
# データ生成　np.linespace ( 始点, 終点, 生成する配列(ndarray)の要素数)
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100) 

# プロット
plt.plot(x, y, label="test")

# 凡例の表示
plt.legend()

# プロットの表示
plt.show()
```

## サインカーブ

```python
# データ生成
x = np.arange(0, 20, 0.1)
y = np.sin(x)

# Figureの初期化
fig = plt.figure(figsize=(12, 8))

# Figure内にAxesを追加() 
ax = fig.add_subplot()
ax.set_title("Sin Curve")
ax.set_label("rad")
ax.plot(x,y)

# 凡例の表示
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)


# プロット表示(設定の反映)
plt.show()
```

## 様々なグラフ

```python
# データ生成
x = np.arange(100)
y = x * np.random.rand(100) 
```

```python
# 折れ線グラフ
plt.plot(x,y)
plt.show()
```

```python
# 棒グラフ
plt.bar(x,y)
plt.show()
```

```python
# ヒストグラム
plt.hist(y, bins=10)
plt.show()
```

```python
# 散布図
plt.scatter(x,y)
plt.show()
```

```python
# 箱ひげ図
plt.boxplot(y)
plt.show()
```
