# 非線形な漸化式

X[n+1] = a * X[n] ( 1 - X[n] ) の振る舞いをグラフで確認

X[0] = 0.2 として a の値を変化させながら・・・

```
import matplotlib.pyplot as plt

# 配列の初期化
n = list(range(100))
X = [0]*100
```

```
X[0] = 0.2
a = 0.6

for i in range(1,100):
  X[i] = a * X[i-1] * (1.0 - X[i-1])

plt.plot(n, X)
plt.show()
```
